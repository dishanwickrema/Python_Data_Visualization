```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Reading CSV
Unemployment = pd.read_csv("../Dataset/Unemployment.csv", encoding="ISO-8859-1")
Unemployment.head(100)
```

```python
Unemployment.set_index('Country', inplace=True)
Unemployment.head()
```

```python
#Counting Rows and Columns
Unemployment.columns
```

```python
# Dropping columns, in order to analyse from 2008 to 2018
Unemployment.drop(['1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998',
       '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007'], axis=1, inplace=True)
Unemployment.head()
```

```python
# Filter Data for Canada, China, Philippines and India
Uemp_compare= Unemployment.loc[["Canada", "China", "India", "Philippines"], :]
Uemp_compare.head()
```

```python
#Transpose the Data to depict the Graph
Uemp_df = Uemp_compare.T
Uemp_df.head(11)
```

```python
Uemp_df.describe()
```

```python
# Plot the line
plt.figure(figsize=(10,5))
plt.title("Unemployment rate among the top immigrant Countires from 2008 to 2018")
plt.plot(Uemp_df)
plt.xlabel('Years')
plt.ylabel('Unemployment Rate')
plt.legend(Uemp_df, loc = 'best')
plt.show()
```
